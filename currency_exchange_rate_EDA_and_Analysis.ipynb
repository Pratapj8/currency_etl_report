## üîÅ Currency Converter

```python
def convert(amount, from_currency, to_currency, df):
    from_currency = from_currency.upper()
    to_currency = to_currency.upper()
    df = df.copy()
    df['base_currency'] = df['base_currency'].str.upper()
    df['target_currency'] = df['target_currency'].str.upper()

    row = df[(df['base_currency'] == from_currency) & (df['target_currency'] == to_currency)]

    if not row.empty:
        rate = row.iloc[0]['rate']
        return amount * rate
    else:
        return f"No rate found for {from_currency} to {to_currency}"
```

```python
# Australian Dollar- AUD
convert(100, 'USD', 'AUD', df)
```

# Detect Arbitrage Opportunities

- Arbitrage occurs when you can exploit price differences in different currency pairs to make a profit. If there‚Äôs a discrepancy in the exchange rates between direct and indirect conversions, it could signal an opportunity.

- We can detect arbitrage by comparing direct exchange rates with indirect exchange rates. For example:

- Convert USD ‚Üí EUR using the direct rate.

- Then, convert USD ‚Üí GBP ‚Üí EUR by going through GBP.

- If the indirect rate differs significantly from the direct rate, there may be an arbitrage opportunity.

```python
# Step 1: Get direct USD ‚Üí EUR rate
usd_eur_direct = filtered_df[(filtered_df['base_currency'] == 'usd') & (filtered_df['target_currency'] == 'eur')]['rate'].iloc[0]

# Step 2: Get indirect USD ‚Üí GBP and GBP ‚Üí EUR rates
usd_gbp = filtered_df[(filtered_df['base_currency'] == 'usd') & (filtered_df['target_currency'] == 'gbp')]['rate'].iloc[0]
gbp_eur = filtered_df[(filtered_df['base_currency'] == 'gbp') & (filtered_df['target_currency'] == 'eur')]['rate'].iloc[0]

# Calculate indirect rate: USD ‚Üí GBP ‚Üí EUR
usd_eur_indirect = usd_gbp * gbp_eur

# Step 3: Compare direct and indirect rates
arbitrage_opportunity = abs(usd_eur_direct - usd_eur_indirect)

print(f"Direct USD ‚Üí EUR Rate: {usd_eur_direct}")
print(f"Indirect USD ‚Üí EUR Rate via GBP: {usd_eur_indirect}")
print(f"Difference (Potential Arbitrage Opportunity): {arbitrage_opportunity}")
```

# Interpretation:
### - The difference between the direct rate and the indirect rate is extremely small (6.47349), which is essentially zero within the precision of most financial data.

# Conclusion:
### - There seems to be no significant arbitrage opportunity for the USD ‚Üí EUR pair via GBP on the given day, as the difference is too small to take advantage of.

# An arbitrage opportunity is like a chance to make money by taking advantage of price differences.
- Imagine you're shopping in two different stores, and one store is selling a product for less than the other. You can buy it cheap from the first store and sell it for a higher price at the second store, making a profit without any risk.
